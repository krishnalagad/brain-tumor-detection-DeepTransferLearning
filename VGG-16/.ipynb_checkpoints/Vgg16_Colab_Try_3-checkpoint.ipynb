```python
train_gen = ImageDataGenerator(
                  rescale=1./255.,
                  horizontal_flip = True,
                   shear_range = 0.2,
                  zoom_range = 0.2
                 )
```

```python
train_generator = train_gen.flow_from_directory(
    directory = train_path, # images data path / folder in which images are there
    color_mode="rgb",
   target_size = (224, 224),
    batch_size = 32,
   class_mode = 'categorical',
    shuffle=True,
    seed=42
)
```

```python
test_gen = ImageDataGenerator(rescale = 1./255,zoom_range = 0.2)
```

```python
valid_generator = test_gen.flow_from_directory(
    directory=test_path,
    target_size=(224, 224),
    color_mode="rgb",
    batch_size=32,
    class_mode="categorical",
    shuffle=True,
    seed=42
)
```

```python
# input shape of one record is (224,224,3) , 16: is the batch size
x,y = next(train_generator)
x.shape
```

```python
y.shape
```

```python
a = train_generator.class_indices
class_names = list(a.keys())  # storing class/breed names in a list
```

```python
class_names
```

```python
base_model = tf.keras.applications.VGG16(
                     include_top=False,
                     weights='imagenet',
                     input_shape=(224,224,3)
                     )
 
base_model.trainable=False
```

```python
 
model = tf.keras.Sequential([
        base_model,  
        tf.keras.layers.BatchNormalization(renorm=True),
        tf.keras.layers.GlobalAveragePooling2D(),
        tf.keras.layers.Dense(512, activation='relu'),
        tf.keras.layers.Dense(256, activation='relu'),
        tf.keras.layers.Dropout(0.5),
        tf.keras.layers.Dense(4, activation='relu'),
        tf.keras.layers.Dense(4, activation='softmax')
    ])
```

```python
model.compile(optimizer='Adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
model.summary()
```

```python
history =model.fit_generator(train_generator,
                    epochs=50,
                    steps_per_epoch=len(train_generator),
                    validation_data=valid_generator,
                    validation_steps=len(valid_generator),
                    
)
```

```python
model.save('D:/Degree/BrainTumor_Impl/VGG-16/'+'vgg16_colab_try_3.h5')
```

```python
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
loss = history.history['loss']
val_loss = history.history['val_loss']
```

```python
# plot results
# accuracy
plt.figure(figsize=(10, 16))
plt.rcParams['figure.figsize'] = [16, 9]
plt.rcParams['font.size'] = 14
plt.rcParams['axes.grid'] = True
plt.rcParams['figure.facecolor'] = 'white'
plt.subplot(2, 1, 1)
plt.plot(acc, label='Training Accuracy')
plt.plot(val_acc, label='Validation Accuracy')
plt.legend(loc='lower right')
plt.ylabel('Accuracy')
plt.title(f'\nTraining and Validation Accuracy. \nTrain Accuracy:{str(acc[-1])}\nValidation Accuracy: {str(val_acc[-1])}')
```

```python
accuracy_score = model.evaluate(valid_generator)
print(accuracy_score)
print("Accuracy: {:.4f}%".format(accuracy_score[1] * 100))
 
print("Loss: ",accuracy_score[0])
```
